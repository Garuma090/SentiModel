```python
# Etapa 1: Instalar as bibliotecas necessárias
!pip install -q transformers

# Etapa 2: Importar bibliotecas
from transformers import pipeline
import numpy as np

# Etapa 3: Criar o analisador de sentimentos
analyzer = pipeline("sentiment-analysis")

# Etapa 4: Entrada do usuário
texto = input("Digite a frase para análise de sentimento: ")

# Etapa 5: Analisar o sentimento
resultado = analyzer(texto)[0]

# Etapa 6: Calcular polaridade média e classificar
label = resultado['label']
score = resultado['score']

# Converte label em pontuação numérica
score_map = {"POSITIVE": 1, "NEGATIVE": -1}
polaridade = score_map.get(label, 0) * score

# Classificação
if polaridade > 0.2:
    classificacao = "POSITIVO (Não agressivo)"
elif polaridade < -0.2:
    classificacao = "NEGATIVO (Agressivo)"
else:
    classificacao = "NEUTRO"

# Exibir resultado
print("\n📝 Resultado da Análise:")
print(f"- Texto: {texto}")
print(f"- Sentimento detectado: {label} ({score:.2f})")
print(f"- Polaridade média: {polaridade:.2f}")
print(f"- Classificação final: {classificacao}")
```
