```python
# Etapa 1: Instalar as bibliotecas necessÃ¡rias
!pip install -q transformers

# Etapa 2: Importar bibliotecas
from transformers import pipeline
import numpy as np

# Etapa 3: Criar o analisador de sentimentos
analyzer = pipeline("sentiment-analysis")

# Etapa 4: Entrada do usuÃ¡rio
texto = input("Digite a frase para anÃ¡lise de sentimento: ")

# Etapa 5: Analisar o sentimento
resultado = analyzer(texto)[0]

# Etapa 6: Calcular polaridade mÃ©dia e classificar
label = resultado['label']
score = resultado['score']

# Converte label em pontuaÃ§Ã£o numÃ©rica
score_map = {"POSITIVE": 1, "NEGATIVE": -1}
polaridade = score_map.get(label, 0) * score

# ClassificaÃ§Ã£o
if polaridade > 0.2:
    classificacao = "POSITIVO (NÃ£o agressivo)"
elif polaridade < -0.2:
    classificacao = "NEGATIVO (Agressivo)"
else:
    classificacao = "NEUTRO"

# Exibir resultado
print("\nðŸ“ Resultado da AnÃ¡lise:")
print(f"- Texto: {texto}")
print(f"- Sentimento detectado: {label} ({score:.2f})")
print(f"- Polaridade mÃ©dia: {polaridade:.2f}")
print(f"- ClassificaÃ§Ã£o final: {classificacao}")
```
